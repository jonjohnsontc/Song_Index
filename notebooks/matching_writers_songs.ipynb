Matching writers to my most recent set of song matches to see what the potential pool of recommended songwriters look like

```python
import json
import re

import numpy as np
import pandas as pd
```

```python
song_list = pd.read_csv('../data/matched_songs_20190209.csv', index_col=0)
writers = pd.read_csv('../data/writers.csv', index_col=0)
cw_lookup = pd.read_csv('../data/cw_lookup.csv', index_col=0)
```

```python
cw_lookup.head()
```

```python
song_list.head()
```

```python
song_list.shape
```

```python
song_list[['CID', 'PID']] = song_list[['CID', 'PID']].astype('int32')
```

```python
combined_song_writer = pd.merge(song_list, cw_lookup, on='CID')
```

```python
combined_song_writer.shape
```

```python
combined_song_writer.head()
```

```python
combined_song_writer = combined_song_writer.merge(writers, on='IPI')
```

```python
combined_song_writer.groupby('IPI').agg('count').sort_values('song_id', ascending=False)
```

```python
writer_song_count = combined_song_writer.groupby(['IPI', 'Writer Name']).agg('count').sort_values('song_id', ascending=False)
```
