```python
song_list = pd.read_csv('/home/jovyan/Song_Index/data/matched_songs_20200103.csv', index_col=0)
writers = pd.read_csv('/home/jovyan/Song_Index/data/writers.csv', index_col=0)
cw_lookup = pd.read_csv('/home/jovyan/Song_Index/data/cw_lookup_20200102.csv')
```

```python
song_list[['CID', 'PID']] = song_list[['CID', 'PID']].astype('int32')
```

```python
combined_song_writer = pd.merge(song_list, cw_lookup, on='CID')
```

```python
combined_song_writer = combined_song_writer.merge(writers, on='WID')
```

```python
combined_song_writer.groupby('IPI').agg('count').sort_values('song_id', ascending=False)
```

```python
writer_song_count = combined_song_writer.groupby(['IPI', 'Writer Name']).agg('count').sort_values('song_id', ascending=False)
```
